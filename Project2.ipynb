```python
import numpy as np
import math
```

```python
matrix = np.zeros((61189,21))
```

```python
classProbMatrix = np.zeros((21,))
```

```python
matrix
```

```python
rowIndex = 1
docIdToLabel = {}
with open('data/train.label', 'r') as trainLabels:
    for line in trainLabels:
        label = int(line.rstrip('\n'))
        docIdToLabel[str(rowIndex)] = label
        classProbMatrix[label] = classProbMatrix[label] + 1
        rowIndex = rowIndex + 1
```

```python
totalTrainingExamples = classProbMatrix.sum()
classProbMatrix = np.log2(classProbMatrix/totalTrainingExamples)
classProbMatrix
```

```python
with open('data/train.data','r') as trainData:
    for line in trainData:
        values = line.rstrip('\n').split(" ")
        label = docIdToLabel[values[0]]
        wordId = int(values[1])
        wordCount = int(values[2])
        matrixValue = matrix[wordId,label]
        newValue = matrixValue + wordCount
        matrix[wordId,label] = newValue
matrix
```

```python
matrix[:,1].sum()
```

```python
vocabSize = 61188.0
```

```python
probMatrix = np.zeros((61189,21))
```

```python
beta = 1.0/vocabSize
```

```python

for v in range(1,21):
    totalWordsInClassV = matrix[:,v].sum()
    if totalWordsInClassV > 1:
        print(totalWordsInClassV)
    for w in range(1,61189):
        wordId = int(w)
        label = int(v)
        countOfWInClassV = matrix[w,v]
        probabilityOfWGivenV = ((countOfWInClassV) + beta)/ (totalWordsInClassV + beta*vocabSize)
        probMatrix[wordId,label] = probabilityOfWGivenV
probMatrix
```

```python
probMatrix = np.log2(probMatrix)
probMatrix
```

```python
for i in range(0,61188):
    value = probMatrix[i,2]
    if value == 0.0:
        print(i,value)
```

```python
testMatrix = np.zeros((7506,61189))
```

```python
rowIndex = 1
docIdToTestLabel = {}
with open('data/test.label', 'r') as testLabels:
    for line in testLabels:
        docIdToTestLabel[str(rowIndex)] = int(line.rstrip('\n'))
        rowIndex = rowIndex + 1
```

```python
with open('data/test.data','r') as testData:   
    for line in testData:
        values = line.rstrip('\n').split(" ")
        label = docIdToTestLabel[values[0]]
        docId = int(values[0])
        wordId = int(values[1])
        wordCount = int(values[2])
        testMatrix[docId,wordId] = wordCount
testMatrix
```

```python
#classifyMatrix = np.zeros((7506,))
classifyMatrix = []
```

```python
i = 1
for n in range(1,7506):
    p = 0.0
    c = 0
    if i == 2000:
        print("processed 2000")
    if i == 4000:
        print("processed 4000")
    if i == 6000:
        print("processed 6000")
    
    for v in range(1,21):
        pv = classProbMatrix[v]
        sum = 0.0
        for w in range(1,61189):
            if testMatrix[n,w] != 0.0:
                sum = sum + testMatrix[n,w] * probMatrix[w,v]
        prob = pv + sum
        #classifyMatrix[n,v] = prob
        if p == 0.0 or prob > p:
            p = prob
            c = v
    classifyMatrix.append(c)
    i = i + 1
 
print("done...")



```
